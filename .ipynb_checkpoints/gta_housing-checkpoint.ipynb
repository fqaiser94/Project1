```python
housing_prices = housing_prices.dropna()
housing_prices.shape
```

Original database was relatively clean. Only 14 rows contained na values 

```python
# Rename the headers to get a clear understanding of the column
housing_prices = housing_prices.rename(columns = {"Area" : "Region",
                                               "Composite_HPI" : "Composite HPI",
                                               "Two_Storey_HPI" : "Two Storey HPI",
                                               "Apartment_HPI" : "Apartment HPI",
                                               "Composite_Benchmark" : "Benchmark Price (Composite)",
                                               "Two_Storey_Benchmark" : "Benchmark Price (Two-Storey)",
                                               "Apartment_Benchmark" : "Benchmark Price (Apartment)"
                                               })
```

```python
housing_prices[['Month','Year']] = housing_prices.Date.str.split(" ", expand=True)
housing_prices.head()
```

```python
clean_housing_prices = housing_prices[['Region','Year','Month','Benchmark Price (Composite)','Benchmark Price (Two-Storey)','Benchmark Price (Apartment)']]

# Formatting data frame
clean_housing_prices.style.format({'Benchmark Price (Composite)':'${:,.2f}',
                                    'Benchmark Price (Two-Storey)':'${:,.2f}',
                                    'Benchmark Price (Apartment)':'${:,.2f}'})
```

```python
group_area_year_month = clean_housing_prices.groupby(['Region','Year','Month']).mean()
group_area_year_month
```

```python
group_area_year = round(clean_housing_prices.groupby(['Region','Year']).mean(),2)
group_area_year
```

```python
group_area_year.reset_index(level=['Year'])
```

```python
group_comp = group_area_year['Benchmark Price (Composite)']
```

```python
group_comp.unstack(level='Region').plot.line(figsize=(20,10), subplots=False)
plt.title('Change in Composite Housing Price per Region')
plt.ylabel('Price ($)')
```

```python
group_twostorey = group_area_year['Benchmark Price (Two-Storey)']
```

```python
group_twostorey.unstack(level='Region').plot.line(figsize=(20,10), subplots=False)
```

```python
group_apart = group_area_year['Benchmark Price (Apartment)']
```

```python
group_apart.unstack(level='Region').plot.line(figsize=(20,10), subplots=False)
```
